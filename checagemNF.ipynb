```python
# Filtrar as contas de compras para estoque por nota fiscal na contabilidade
# contabilidade_compras = df[df['CFOP'] == 1102]['Nota'].tolist()
contabilidade_compras = df['Nota'].tolist()  ###df[df['CFOP'] == 1102]['Nota'].tolist()
print(contabilidade_compras)

# Extrair as notas fiscais da escrita fiscal
escrita_fiscal_compras = df2['Nota'].tolist()
print(escrita_fiscal_compras)

```

```python
# Verificar se as notas fiscais da contabilidade estão presentes na escrita fiscal
notas_faltando = [nf for nf in contabilidade_compras if nf not in escrita_fiscal_compras]

if notas_faltando:
    print("Notas fiscais faltando na escrita fiscal:")
    for nf in notas_faltando:
        print(nf)
else:
    print("Todas as notas fiscais da contabilidade estão presentes na escrita fiscal")
```
